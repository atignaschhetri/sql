1. CREATE (Insert Data)
It is used to add new record into table.

```python
%%sql
CREATE TABLE if not exists students(
    id INT NOT NULL,
    people_name VARCHAR(20),
    birth_date DATE NOT NULL,
    phone VARCHAR(15),
    PRIMARY KEY (id)
);
```

2.INSERT 
Used to insert Data into table.

```python
%%sql
INSERT INTO students (id, people_name, birth_date, phone)
VALUES
(1, 'Ram', '1985-08-23', '9123456780'),
(2, 'leo', '1992-11-10', '9988776655'),
(3, 'Brown', '1998-03-17', '9090909090'),
(4, 'john', '1980-12-01', '9008007006');
```

3.READ
It is used to fetch  from the table.

```python
%%sql 
select * from students;
```

```python
%%sql
select * from students where people_name='leo';;
```

4.UPDATE or Modify Data
It is used to modify or update the data into table

```python
%%sql
update students
set people_name = 'Hari'
where id = 4;
```

```python
%%sql
select * from students where birth_date > '1990-01-01';
```

```python
%%sql
select * from students where birth_date > '1985-08-23';
```
